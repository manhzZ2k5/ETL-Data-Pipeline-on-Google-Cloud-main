```python
import csv
Link_list = []
data_list = []
with open('mogi_links.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['link'])

    for page in range(1, 5):
        url = f"{base_url}?&cp={page}"
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")

        links = soup.find_all("a", class_="link-overlay")
        for link in links:
            href = link.get("href")
            if href:
                writer.writerow([href])
                Link_list.append(href)

for link in Link_list:
    res = requests.get(link)
    if res.status_code == 200:
        data = {}
        sop = BeautifulSoup(res.text,"html.parser")
        infos = sop.find_all("div", class_ = "info-attr clearfix")
        price = sop.find('div',class_ = "price")
        address = sop.find('div',class_ = "address")
        data['price'] = price.get_text(strip = True)
        data['address'] = address.get_text(strip = True)
        for info in infos:
            span = info.find_all('span')
            if len(span) >= 2:
                key = span[0].get_text(strip=True)
                value = span[1].get_text(strip=True)
                data[key] = value
        data_list.append(data)
        time.sleep(1)

fieldnames = set()
for d in data_list:
    fieldnames.update(d.keys())
fieldnames = list(fieldnames)

with open('batdongsan.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data_list)

print("Đã lưu dữ liệu vào file batdongsan.csv")
```

```python
# Đọc file và kiểm tra duplicates
df = pd.read_csv("batdongsan.csv")
df_no_duplicates = df.drop_duplicates()

# In ra số lượng rows trước và sau khi xử lý
print(f"Số dòng ban đầu: {len(df)}")
print(f"Số dòng sau khi loại bỏ trùng lặp: {len(df_no_duplicates)}")
print(f"Số dòng trùng lặp đã xóa: {len(df) - len(df_no_duplicates)}")

# Kiểm tra tỷ lệ trùng lặp
duplicate_rate = ((len(df) - len(df_no_duplicates)) / len(df)) * 100
print(f"Tỷ lệ trùng lặp: {duplicate_rate:.2f}%")
```
